```python
a= rm.random()
print(a)
b = rm.randint(10,20) # 10~20
print(b)
c = rm.randrange(10,20,2)# 10,12,14,16,18,20
print(c)
```

```python
t1 = [1,2,3,4,5]
rm.shuffle(t1) # random으로 섞음
print(t1)
a = rm.choice(t1) # 임의의 항목을 선택
print(a)
```

```python
a = 10
b = sts.math.sqrt(a)
print(b)
t1 = [2,3,5,7,11]
a = sts.mean(t1) # 평균값
print(a)
a = sts.median(t1) # 중앙값
print(a)
```

```python
t1 = [1,2,3]
# 얕은복제
a = cp.copy(t1)
print(t1 == a)
print(t1 is a)
print("===========")

a[0] = 999
print("999추가후")
print(t1)
print(a)
print("===========")

c= [1,2,3,[4,5]]
d = cp.copy(c)
print(c)
print(d)
print("===========")
d[-1].append(6) # d만 바뀐것이 아닌 c도 변경이 되버림
print(c)
print(d)
print("===========")

# 깊은복제
b = cp.deepcopy(t1)
print(t1 is b)
print(t1)

print("===========")
b[0] = 777
print("777추가후")
print(b)
print("===========")
c= [1,2,3,[4,5]]
d = cp.deepcopy(c)
print(c)
print(d)
print("===========")
d[-1].append(6)
print(c)
print(d)
```
