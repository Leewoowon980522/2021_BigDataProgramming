```python
# 예외 내용을 변수에 대체하여 처리. as 변수명을 사용하여 처리할수 있다
try : 
    f = open("nofile.txt")
    f.close()
    
    print(1+2+c)
    
    print([1,2,3][3])
except IndexError as e1:
    print(e1)
except FileNotFoundError as e2:
    print(e2)
except Exception as e:
    print(e)
```

```python
# 예외 내용을 변수에 대체하여 처리. as 변수명을 사용하여 처리할수 있다
try : 
    f = open("test.txt")
    f.close()
    
    print(1+2+c) # 없는 변수 c를 사용했기 때문에 예외 발생
    
    print([1,2,3][3])
except IndexError as e1:
    print(e1)
except FileNotFoundError as e2:
    print(e2)
except Exception as e:
    print(e)
```

```python
# 예외처리의 추가 구문,else
# 예외가 없다면 else구문이 실행이 된다
try : 
    f = open("test.txt")
    f.close()
    
    print(1+2+3)
    
    print([1,2,3][0])
except Exception as e:
    print(e)
else:
    print("예외 발생 없이 try 블록이 모두 수행되었습니다")
```

```python
# try-except에 이어지는 finally는 예외 발생여부와 상관없이 무조건 실행되는 블록
try : 
    f = open("nofile.txt")
    f.close()
    
    print(1+2+3)
    
    print([1,2,3][0])
except Exception as e:
    print(e)
finally:
    print("예외 여부와 무관하게 이 블록은 항상 실행됩니다")
```

```python
# 예외 처리의 예1
try:
    v = int(input("정수를 입력하세요: "))
    if v>0:
        print("양수입니다")
    elif v<0:
        print("음수입니다")
    else:
        print("0입니다")

except ValueError as ve:
    print("정수로 변환할수 없는 값입니다.")
    print(ve)
```

```python
# 예외 처리의 예2
while True:
    try:
        v = int(input("정수를 입력하세요: "))
        if v>0:
            print("양수입니다")
        elif v<0:
            print("음수입니다")
        else:
            print("0입니다")
        break
    except ValueError as ve:
        print("정수로 변환할수 없는 값입니다.")
        print(ve)
```

```python
# 예외 처리의 예3
def func1(a,b):
    return a/b
def func2(c):
    return c[-1]

try:
    r1 = func1(5,0) # 0으로 나눌수 없어서 예외발생
    r2 = func2({1,2,3})
except Exception as e:
    print(e)
    
try:
    r1 = func1(123,456)
    r2 = func2({1,2,3}) # set은 순서가 없어 인덱싱이 불가 예외 발생
except Exception as e:
    print(e)

    
# //===== 정상 작동 =====//
try:
    r1 = func1(123,456)
    r2 = func2([1,2,3])
except Exception as e:
    print(e)
```
